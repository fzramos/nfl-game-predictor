```python
from sportsreference.nfl.boxscore import Boxscores, Boxscore
import pandas as pd
```

```python
# Arg1: Weeek, Arg2: Year
Boxscores(1,2020).games
# this just gives you the scores, winner, and home team.
# Critically, it also gives the game "boxscore" id.
# each game has a "boxscore" id value which you can use to get more game details|
```

```python
# Get stats for a specific game
game_str = Boxscores(1,2020).games['1-2020'][0]['boxscore']
# Each game has a "boxscore" id value 
# you can use that id to get more details about the game by using the Boxscore() function as seen below
game_stats = Boxscore(game_str)
game_stats.dataframe
```
